```python
import matplotlib.pyplot as plt 
import matplotlib.patches as patches
import time
```

CREATING ENVIRONMENT

```python
#  0 -> Dirty, 1 -> Clean
environment={
    "Room1":1,
    "Room2":0,
    "Room3":1,
    "Room4":1
}

room_position={
    "Room1":(0,1),
    "Room2":(1,1),
    "Room3":(0,0),
    "Room4":(1,0)
}
rooms=list(environment.keys())

def draw_environment(env,agent_index,step):
    fig,ax=plt.subplots()
    ax.set_xlim(0,2)
    ax.set_ylim(0,2)
    # ax.set_xticks([])   # Show/Hide x/y readings
    # ax.set_yticks([])
    ax.set_title(f"Step {step} - Agent in {rooms[agent_index]}")
    
    for room,pos in room_position.items():
        x,y=pos
        color='red' if env[room]==0 else 'green'
        #patches.Rectangle(Bottom-left,width,height,bg-color,edge-color)  color=facecolor+edgecolor
        rect=patches.Rectangle((x,y),1,1,facecolor=color,edgecolor='black')
        ax.add_patch(rect)
        ax.text(x+0.5,y+0.5,room,ha='center',va='center',color='white')
    
    #Drawing Agent
    agent_x,agent_y=room_position[rooms[agent_index]]
    #patches.Circle(Bottom-left,radius,bg-color,edge-color)
    agent_patch=patches.Circle((agent_x+0.5,agent_y+0.5),0.1,facecolor='blue',edgecolor='white')
    ax.add_patch(agent_patch)
    
    # wait for 1 second
    plt.pause(1) 
    plt.close()
```

CREATING REFLEX AGENT FUNCTION

```python
def reflex_agent(state):
    if state==0:
        return 1
    return -1
```

RUNNING SIMULATION (Using IPython)

```python
from IPython.display import clear_output # need to import this for same line frames
steps=8
agent_index=0
for step in range(steps):
    current_room = rooms[agent_index]
    state = environment[current_room]
    action = reflex_agent(state)

    clear_output(wait=True) # frame will see as a linear
    draw_environment(environment, agent_index, step)

    if action == 1:
        environment[current_room] = 1
    else:
        agent_index = (agent_index + 1) % len(rooms)
print("Simulation Completed")
```

RUNNING SUMULATION (AS FRAME BY FAME OUTPUT)

```python
plt.ion() # Interation mode on  (require  only in .py files)
steps=9
agent_index=0
for step in range(steps):
    current_room=rooms[agent_index]
    state=environment[current_room]
    action=reflex_agent(state)
    
    draw_environment(environment,agent_index,step)
    
    if(action==1): # if clean
        environment[current_room]=1
    else:
        agent_index=(agent_index+1)%len(rooms)

plt.ioff()
print("Simulation Completed")
    
```
